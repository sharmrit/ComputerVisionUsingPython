```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def detectCorners(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Convert Gray Scale Image to Float Values
    gray = np.float32(image)
    
    # Corner Harris Detection
    res = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)

    # result is dilated for marking the corners
    # this is just so we can plot out the points on the image shown
    res = cv2.dilate(res,None)

    # Threshold for an optimal value, it may vary depending on the image.
    img[res>0.05*res.max()]=[255,0,0]
    
    plt.imshow(img)
```

```python
img = cv2.imread('./data/brickWall.jpg')
detectCorners(img)
```
