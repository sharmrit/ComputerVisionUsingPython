```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
    

def detect_eyes(img):
    
    eyes_img = img.copy()
    roi = img.copy()
    eye_cascadeclassfier = cv2.CascadeClassifier('./data/haarcascades/haarcascade_eye.xml')
    eye_rects = eye_cascadeclassfier.detectMultiScale(eyes_img,scaleFactor=1.3, minNeighbors=3) 
    
    for (x,y,w,h) in eye_rects:
        cv2.circle(eyes_img, (int(x+0.5*w),int(y+0.5*h)), int(0.5*np.sqrt(w*w+h*h)), (0,255,0), 3)
  
    return eyes_img

```

```python
img = cv2.imread('./data/actor.jpg')
display(detect_eyes(img))
```
